### Reorganizing some of the bean types into single category as they are subtype of same bean.

```python
# Checking unique values of bean type
df_cleaned.Bean_type.unique()
```

```python
# Replacing the same category beans as per name 
df_cleaned['Bean_type'].replace('Forastero (Arriba) ASSS', 'Forastero',inplace=True)
df_cleaned['Bean_type'].replace('Forastero (Arriba) ASS', 'Forastero',inplace=True)
df_cleaned['Bean_type'].replace('Forastero (Arriba)', 'Forastero',inplace=True)
df_cleaned['Bean_type'].replace('Forastero (Nacional)', 'Forastero',inplace=True)
df_cleaned['Bean_type'].replace('Criollo, +','Criollo',inplace=True)
df_cleaned['Bean_type'].replace('Blend-Forastero,Criollo','Blend',inplace=True)
df_cleaned['Bean_type'].replace('Forastero(Arriba, CCN)','Forastero',inplace=True)
df_cleaned['Bean_type'].replace('Forastero (Amelonado)','Forastero',inplace=True)
df_cleaned['Bean_type'].replace('Trinitario, Nacional','Trinitario',inplace=True)
df_cleaned['Bean_type'].replace('Trinitario (Amelonado)','Trinitario',inplace=True)
df_cleaned['Bean_type'].replace('Trinitario, TCGA','Trinitario',inplace=True)
df_cleaned['Bean_type'].replace('Criollo (Amarru)','Criollo',inplace=True)
df_cleaned['Bean_type'].replace('Criollo, Trinitario','Blend',inplace=True)
df_cleaned['Bean_type'].replace('Criollo (Porcelana)','Criollo',inplace=True)
df_cleaned['Bean_type'].replace('Trinitario (85% Criollo)','Blend',inplace=True)
df_cleaned['Bean_type'].replace('Forastero (Catongo)','Forastero',inplace=True)
df_cleaned['Bean_type'].replace('Forastero (Parazinho)','Forastero',inplace=True)
df_cleaned['Bean_type'].replace('Trinitario, Criollo','Blend',inplace=True)
df_cleaned['Bean_type'].replace('Criollo (Ocumare)','Criollo',inplace=True)
df_cleaned['Bean_type'].replace('Criollo (Ocumare 61)','Criollo',inplace=True)
df_cleaned['Bean_type'].replace('Criollo (Ocumare 77)','Criollo',inplace=True)
df_cleaned['Bean_type'].replace('Criollo (Ocumare 67)','Criollo',inplace=True)
df_cleaned['Bean_type'].replace('Criollo (Wild)','Criollo',inplace=True)
df_cleaned['Bean_type'].replace('Trinitario, Forastero','Blend',inplace=True)
df_cleaned['Bean_type'].replace('Trinitario (Scavina)','Trinitario',inplace=True)
df_cleaned['Bean_type'].replace('Criollo, Forastero','Blend',inplace=True)
df_cleaned['Bean_type'].replace('Forastero, Trinitario','Blend',inplace=True)
df_cleaned.replace('\xa0','Unkown',inplace=True)
df_cleaned['Bean_type'].replace(np.nan,'Unkown',inplace=True)

print (df_cleaned['Bean_type'].unique())
```

```python
# Assigning number categories to the beans for further analysis
def bean_type(bean):
    if   bean == 'Criollo':
        return 1
    elif bean == 'Trinitario':
        return 2
    elif bean == 'Forastero':
        return 3
    elif bean =='Blend':
        return 4
    elif bean =='Unkown':
        return 5
    else:
        return 6

df_cleaned['Bean_type']=df_cleaned['Bean_type'].map(bean_type)
df_cleaned['Bean_type'].unique()
```

## CORRELATION

### Q3.Create a correlation matrix heatmap (colored)

```python
# Creating a correlation matrix
df_cleaned_sub.corr()
```

```python
# Create a subplot with matplotlib
f,ax = plt.subplots(figsize=(10,10))

# Create the correlation heatmap in seaborn by applying a heatmap onto the correlation matrix and the subplots defined above.
corr = sns.heatmap(df_cleaned_sub.corr(), annot = True, ax = ax) # The `annot` argument allows the plot to 
#place the correlation coefficients onto the heatmap.
```

### From the heatmap above,following observations can be made-
### 1.There seems to be a very weak positive correlation (0.038) between the review year and the cocoa percentage. This suggests that there might be a slight tendency for cocoa percentage to increase slightly over the years, but the correlation is too weak to draw any strong conclusions.
### 2. Cocoa Percentage vs. Rating: There is a moderate negative correlation (-0.16) between cocoa percentage and rating. This indicates that as the cocoa percentage increases, the rating tends to decrease. This could imply that higher cocoa percentages might not be as palatable to all consumers, resulting in lower ratings on average.
### 3.Review Year vs. Rating: There is very weak positive correlation (0.1) between the review year and the rating. This suggests that over the years, there might be a slight tendency for ratings to increase slightly. This could be due to factors such as improvements in chocolate-making techniques or changes in consumer preferences.
### Correlation does not imply causation. Further analysis and consideration of other factors are necessary to understand the relationships fully.
